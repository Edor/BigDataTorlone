```python
import csv
import time
```

```python
rdd3 = sc.textFile("hdfs://localhost:9000/user/emo/input/Bigreviews.csv")
rdd3 = rdd3.mapPartitions(lambda x: csv.reader(x))
```

```python
startTime = time.time()

header3 = rdd3.first()
otherrdd = rdd3.filter(lambda line: line != header3) \
    .map(lambda line: (line[2], line[1]))
tmpRdd3 = rdd3.filter(lambda line: line != header3) \
    .map(lambda line: (line[2], line[1])) \
    .leftOuterJoin(otherrdd) \
    .filter(lambda line: line[1][0] != line[1][1]) \
    .map(lambda line: (line[1], 1)) \
    .reduceByKey(lambda a,b:a+b) \
    .sortBy(lambda x: x[0][0])
    
tmpRdd3.saveAsTextFile("hdfs://localhost:9000/user/emo/output/sparkbig3/")
finishTime = time.time()
print(finishTime - startTime)
```
